# Logistic Regression 

$$ z = A + B \cdot X_1 + C \cdot X_2$$
$$ P = \sigma(z) = \frac{1}{1+e^{-z}}$$
Loss function is
$$ L = -y \cdot \ln(P) - (1-y) \cdot \ln(1-P)$$
$$ y\in \{0, 1\}$$

## Task 1 

Derive 
$$\sigma'(z) = f(\sigma(z))=f(z)$$
$$\ln(\sigma(z)) = f(z)$$
$$\ln(1-\sigma(z)) = f(z)$$

## Task 2 

Derive
$$ \frac{\partial L}{\partial A}$$
$$ \frac{\partial L}{\partial B}$$
$$ \frac{\partial L}{\partial C}$$

## Task 3

Estimate the mean value of $P = \sigma(z)$ $\text{if } z = -6 + 0.07 \cdot X_1 + 0.5 \cdot X_2$
$$ X_1 \sim N(0, 1)$$
